```python
import numpy as np
import pandas as pd
import random
import altair as alt
```

Simulation for low cardinality estimation.

```python
bucket_size = 4096
```

```python
buckets = np.zeros(bucket_size)
num_logs = 50000

buckets_taken = np.zeros(num_logs)

for i in range(num_logs):
    bucket_index = hash(str(i)) % bucket_size
    buckets[bucket_index] = 1
    buckets_taken[i] = np.count_nonzero(buckets)

source = pd.DataFrame({
  'Number of Logs': np.arange(0, num_logs, 10),
  'buckets taken': buckets_taken[::10]
})    
reference_source = pd.DataFrame({
  'Number of Logs': np.arange(0, num_logs, 10),
  'buckets taken': (bucket_size - bucket_size * np.exp(-np.arange(0, num_logs, 10) / bucket_size))
})
```

```python
chart = alt.Chart(source).mark_line().encode(
    x= 'Number of Logs',
    y='buckets taken', 
)
reference_chart = alt.Chart(reference_source).mark_line(color='gray').encode(
    x= 'Number of Logs',
    y='buckets taken', 
)
# They should nearly overlap
reference_chart + chart
```

```python
diff = pd.DataFrame({
  'Number of Logs': np.arange(0, num_logs, 10),
  'Buckets difference': (source - reference_source)['buckets taken'] / np.arange(0, num_logs, 10)
}) 
alt.Chart(diff).mark_line().encode(
    x= 'Number of Logs',
    y='Buckets difference', 
)
```
